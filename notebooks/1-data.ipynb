```julia
using GynC
using Plots
```

```julia
plots=[]

factors = [1,1,3.6712,1]

for spec = 1:4
  specind = GynC.measuredinds[spec]
  p=plot(title=GynC.speciesnames[specind])#, size=plotsize)
  for i=1:45
      d=Lausanne(i)
      scatter!(0:1:30, d.data[:,spec], label="", ms=5)
  end
    
  ref = GynC.referencesolution(1/10)[:,specind] * factors[spec]
  plot!(0:1/10:30, ref, label="reference", c=:black, width=3, legend=:topleft)
  push!(plots, p)
end

plot(layout=4, plots..., size=(800,500))
```
