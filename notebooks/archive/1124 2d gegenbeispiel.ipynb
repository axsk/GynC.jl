```julia
using GynC, Plots
```

```julia
include("../src/eb/plot.jl")
import Plots.scatter

function scatter(ys::Vector{Vector{Float64}}; kwargs...)
  m = hcat(ys...)
  scatter(m[1,:], m[2,:]; kwargs...)
end
```

```julia
sigma = 3
nx = 50
nz = 300
m = Federn.federmodel(nx, 0, nz, sigma)
w0 = Federn.wbeta(m.xs, maximum(m.xs));
```

```julia
niter = 100
h=1
wzs = GynC.mple(m, w0, niter, 1, h)
wz = wzs[end];
```

```julia
plot(map(norm, wzs[2:end]-wzs[1:end-1]))
```

```julia
scatter(m.ys)
```

```julia
plot(m.xs, wz)
```

```julia
scatter(m.ys, ms=wz*500, lims=(-15,15), size=(600,600))
```

```julia
plot_kde(m.ys, wz, bandwidth=sigma, legend = false)
```
