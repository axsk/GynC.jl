## chainanalysis

```julia
using GynC
using Plots
blas_set_num_threads(1)

s = load("pL3s0.1r0.001t1.jld")
```

```julia
# jumpsize plot
p = plot()
for i = [1,8,50,90,110]
    u=unique(s.samples[:,i]) 
    
    x = map(x->first(findin(s.samples[:,i], x)), u)
    v = (u[2:end] - u[1:end-1]) ./ s.samples[1,i]
    
    scatter!(x, v, label=i, m=(3,stroke(0)), xlim=(0,size(s.samples, 1)))
end
p
```
