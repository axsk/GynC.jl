```julia
addprocs(10)
using GynC
using Plots; gr()
using Interact
#include("../src/eb/plot.jl");
```

```julia
rho_std = 1
nx = 500
ndata = 50000
zmult = 600

srand(1)
m  = Federn.federmodel(nx, ndata, zmult, rho_std);
w0 = Federn.betaprior(m);
```

```julia
h = 1
#reg = 1 - 1/20
reg = .96
niter = 100

ws   = @time GynC.mple(m, w0, niter, reg, h)
emws = @time GynC.em(m, w0, niter);

hmax = GynC.mple(m, w0, niter, 1, h);
```

```julia
# comparison plots
labels = ["MPLE" "NPMLE"];

#final posterior
plot(m.xs, [ws[end] emws[end] hmax[end]], labels=hcat(labels, "hmax")) |> display

hz(w) = GynC.hz(m, w)
ll(w) = GynC.logl(m, w)

# objective iterations
plot(hz.([ws emws]), yaxis="H_z",  labels=labels) |> display
plot(ll.([ws emws]), yaxis="logL", labels=labels) |> display

#plot((1-reg)*(ll.(ws)) + reg*hz.(ws), yaxis="obj (reg=$reg)") |> display
```
