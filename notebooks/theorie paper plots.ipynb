```julia
addprocs(10)
using GynC
using Plots; gr()
using Interact
#include("../src/eb/plot.jl");
```

```julia
rho_std = 1
rho_y = 5
nx = 100
ndata = 20
zmult = 300

srand(1)
m  = Federn.federmodel(nx, ndata, zmult, rho_std);
w0 = Federn.betaprior(m);

my = deepcopy(m);
my.zs = m.ys
my.zsampledistr = Distributions.MvNormal(2, rho_y);
```

```julia
h = 0.1
reg = 1 - 1/ndata
#reg = .92
niter = 200;
```

```julia
hzmax = @time GynC.mple(m,  w0, niter, 1, h);
hymax = @time GynC.mple(my, w0, niter, 1, h);
```

```julia
plot(m.xs, [hzmax[end] hymax[end]], labels=["hz" "hy"]) |> display
```

```julia
ws   = @time GynC.mple(m, w0, niter, reg, h)
wsy  = @time GynC.mple(my, w0, niter, reg, h)
#emws = @time GynC.em(m, w0, niter);
```

```julia
plot(m.xs, [ws[end] wsy[end]], labels=["ws" "wsy"]) |> display
```
