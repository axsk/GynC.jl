```julia
using ClusterManagers
addprocs(SlurmManager(20))
using GynC
```

```julia
datadir = "../data/batch"
selector = "0.001"
window   = 400_000:500_000;
```

```julia
files = filter(x->contains(x, selector), readdir(datadir));

@time samplings = GynC.Sampling[GynC.load(joinpath(datadir, f)) for f in files]

haswindow = map(s->size(s.samples, 1), samplings) .>= maximum(window)

samplings = map(s->s[window,:], samplings[haswindow]);
```

```julia
println(length(samplings), " samplings a ", size(samplings[1].samples, 1), " samples each")

np = length(workers())
uniques = map(s->s.samples[:,1]|>unique|>length, samplings) |> sum
println("total unique samples: ", uniques)
println("estimated time on $np cores: ", uniques * length(samplings) / 100 / 3600 / np, " hours")
```

```julia
@time wc = WeightedChain(samplings);
```

```julia
Plots.bar(GynC.weightedhist(wc.samples[:,8]|>vec, wc.weights, 10)...)
```

```julia
using JLD
JLD.save(joinpath(datadir, "allchains.jld"), "wc", wc)
```
