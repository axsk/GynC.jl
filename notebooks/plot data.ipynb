```julia
using GynC
using Plots
```

```julia
plotsize=(600, 400*0.8)
for spec = 1:4
  specind = GynC.measuredinds[spec]
  p=plot(title=GynC.speciesnames[specind], size=plotsize)
  for i=1:55
      d=Lausanne(i)
      scatter!(d.data[:,spec], label="")
  end
    
  ref = GynC.referencesolution(1/10)[:,specind]
  plot!(0:1/10:30, ref, label="reference", c=:black, width=4)
  display(p)
end
```
