## Optimization

```julia
using NLopt

eqcalgs = [:GN_ISRES, :LN_COBYLA, :LN_BOBYQA, :LN_NEWUOA_BOUND, :LN_PRAXIS, :LN_NELDERMEAD, :LN_SBPLX]
opt = Opt(eqcalgs[6], K)

lower_bounds!(opt, zeros(K))
upper_bounds!(opt, ones(K))

#equality_constraint!(opt, (w,g)-> sum(w) - 1)
#max_objective!(opt, (w,g) -> o(w) - ((sum(w)-1)*10000)^20)
max_objective!(opt, (w,g) -> o(GynC.projectsimplex(w)))

maxeval!(opt, 1000)

@time v,w = optimize!(opt, ones(K)/K)
@show sum(w)
#@assert sum(w)-1 < 0.001
v
```
