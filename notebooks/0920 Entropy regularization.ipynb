## Optimization

```julia
normedPL(w) = o(w/sum(w))

res = ForwardDiff.GradientResult(rand(K))

function obj(w,g)
    ForwardDiff.gradient!(res, normedPL, w)
    if length(g) > 0
        g[:] = ForwardDiff.gradient(res)
    end
    ForwardDiff.value(res)
end


using NLopt

derfree = [:GN_ISRES, :LN_COBYLA, :LN_BOBYQA, :LN_NEWUOA_BOUND, :LN_PRAXIS, :LN_NELDERMEAD, :LN_SBPLX]
ld = [:LD_LBFGS, :LD_MMA, :LD_SLSQP]
opt = Opt(ld[1], K)

lower_bounds!(opt, zeros(K))
upper_bounds!(opt, ones(K))

#equality_constraint!(opt, (w,g)-> sum(w) - 1)
#max_objective!(opt, (w,g) -> o(w) - ((sum(w)-1)*10000)^20)

max_objective!(opt, obj)

maxeval!(opt, 5000)

@time v,w = optimize!(opt, ones(K)/K)
@show sum(w)
v, w[1:10:end]
```
