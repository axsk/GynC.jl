```julia
using GynC
```

```julia
srand(1)
sim = GynC.mcmc(GynC.ModelConfig(GynC.Pfizer(1)), 3000, relprop = 0.1)
```

```julia
for s = 1:4 
    GynC.plot_residuum(sim[1:50:end,:,:], s, linewidth=0.7, alpha=0.9)
    display() 
end
```

```julia
map(18:20) do i
    map(1:4) do s
        data = fill(NaN, 4, 31)
        data[s,:] = sim.model[:data].value[s,:]
        GynC.llh(data, GynC.allparms(sim[i:i,:parms,:].value|>vec), sim[i:i,:y0,:].value|>vec, 0.1)
    end
end
```
