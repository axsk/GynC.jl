```julia
function sample(wc::WeightedChain)
    m = maximum(wc.weights)
    while true
        i = rand(1:size(wc.samples, 1))
        wc.weights[i] / m > rand() && return wc.samples[i,:] |> vec
    end
end
```

```julia
import GynC: sampletoparms, gync, WeightedChain, measuredinds, Sampling

function layersols(c::WeightedChain, n::Integer, species=1)
    t = collect(0:0.3:30.)
    species = measuredinds[species]
    for i in 1:n
        p, y0 = sampletoparms(sample(c))
        sol = gync(y0, p, t)
        plot!(t, sol[species, :]|>vec, color=:blue, legend=:none, alpha=0.1)
    end
end

function plotsample(sample::Vector, species::Integer; plotargs...)
    p, y0 = sampletoparms(sample)
    sol = gync(y0, p, t)
    plot!(t, sol[species, :] |> vec; plotargs...)
end

function plotdata(sim::Sampling, species=1)
    patientdata = sim.config.data
    plot!(patientdata[species,:]|>vec, marker=:x)
end

p=plot()
species = 2

for i in 1:100
    plotsample(sample(c), measuredinds[species]; legend=:none, alpha=0.1, color=:blue)
end

for i in 1:3
    plotdata(sims[i], species)
end

display(p)
```
